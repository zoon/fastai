```python
%matplotlib inline
from fastai.imports import *
from sklearn.ensemble import RandomForestRegressor
```

```python
x = np.linspace(0, 1)[..., np.newaxis]
y = (x + np.random.uniform(-.2, .2, x.shape)).ravel()
print(x.shape, y.shape)

x_trn, x_val = x[:40], x[40:]
y_trn, y_val = y[:40], y[40:]
```

```python
plt.scatter(x, y);
```

```python
m = RandomForestRegressor(n_estimators=10).fit(x_trn, y_trn)
```

```python
plt.scatter(y_trn, m.predict(x_trn))
```

```python
plt.scatter(y_val, m.predict(x_val))
```
